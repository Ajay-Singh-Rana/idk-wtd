```python
from manim import *
config.media_embed = True
```

```python
%%manim -v WARNING --disable_caching -ql CreateCircle

config.media_width = "75%"
config.media_embed = True

class CreateCircle(Scene):
    def construct(self):
        circle = Circle()
        circle.set_fill(BLUE,opacity = 0.6)
        self.play(Create(circle))
```

```python
%%manim -v WARNING --disable_caching -ql TransformCircleToRectangle

class TransformCircleToRectangle(Scene):
    def construct(self):
        circle = Circle()
        circle.set_fill(YELLOW, opacity = 0.3)
        
        rect = Rectangle()
        rect.set_fill(YELLOW, opacity = 0.3)
        
        self.play(Create(circle))
        self.play(Transform(circle ,rect))
        self.play(FadeOut(circle))
```

```python
%%manim -v WARNING --disable_caching -ql Positioning

class Positioning(Scene):
    def construct(self):
        circle = Circle()
        self.play(Create(circle))
        
        square = Square()
        self.play(Create(square))
        self.play(square.animate().next_to(circle,RIGHT,buff=))
        self.play(circle.animate().set_fill(BLUE, opacity = 0.5))
```
